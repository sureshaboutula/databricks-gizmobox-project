# Create Customer Order Summary
  1. Join the tables silver_customers, silver_addresses, silver_orders
  2. Retrieve the latest Address of the Customer
  3. Calculate the following Values
-           total_orders
-           total_items_ordered 
-           total_order_amount

![CircuitBox_data_model.png](./CircuitBox_data_model.png "CircuitBox_data_model.png")

### Customer Order Summary Requirement

![customer_order_summary.png](./customer_order_summary.png "customer_order_summary.png")

```sql
create or refresh materialized view gold_customer_order_summary
as
select c.customer_id, c.customer_name, c.date_of_birth, c.telephone, c.email,
       a.address_line_1, a.city, a.state, a.postcode,
       count(distinct order_id) as total_orders,
       sum(o.item_quantity) as total_items_ordered,
       sum(o.item_quantity * item_price) as total_order_amount
from live.silver_customers c
join live.silver_addresses a on c.customer_id = a.customer_id
join live.silver_orders o on c.customer_id = o.customer_id
where a.__END_AT is null
group by ALL;
```
