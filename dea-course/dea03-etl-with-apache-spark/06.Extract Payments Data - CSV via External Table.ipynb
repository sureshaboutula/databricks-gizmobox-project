### 1. List the files from the Payments Folder

```sql
%fs ls 'abfss://gizmobox@deacourseextdldev.dfs.core.windows.net/landing/external_data/payments'
```

### 2. Create External Table

```sql
create table if not exists gizmobox.bronze.payments
  (payment_id integer, order_id integer, payment_timestamp timestamp, payment_status integer, payment_method string)
using CSV
options(
  header = "true",
  delimiter = ","
)
location 'abfss://gizmobox@deacourseextdldev.dfs.core.windows.net/landing/external_data/payments'
```

```sql
select * from gizmobox.bronze.payments;
```

```sql
describe extended gizmobox.bronze.payments;
```

### 3. Demonstrate the effect of adding/updating/deleting files.

```sql
-- we have 87 records in the table before making any changes to the payments folder
-- Now delete one of the files and run select * statement again and check the number of records

select * from gizmobox.bronze.payments;

-- we are getting error saying that the deleted file is does not exist
-- We can fix this by refreshing the metadata in the unity catalog - refresh table command
```

```sql
refresh table gizmobox.bronze.payments;
```

```sql
select * from gizmobox.bronze.payments;
-- now we have 60 rows only
```

```sql
-- Now add the deleted file back and run select * statement again and check the number of records
select * from gizmobox.bronze.payments;
-- still we see 60 records instead of 87. Thats because the metadata has only 3 files. So we need to refresh the metadata.
```

```sql
refresh table gizmobox.bronze.payments;
```

```sql
select * from gizmobox.bronze.payments;
-- now we have 87 records
```

### 4. Demonstrate the effect of Dropping the table

```sql
drop table if exists gizmobox.bronze.payments
```
