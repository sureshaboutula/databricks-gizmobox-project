# Configure Access to Azure SQL Database to extract Refunds data via Lakehouse Federation
1. Create connection - azure_sql_gizmobox_db_conn
2. Create foreign catalog - azure_sql_gizmobox_db_catalog 

### 1. Create connection - azure_sql_gizmobox_db_conn

```sql
create connection azure_sql_gizmobox_db_conn type sqlserver
options(
  host 'gizmobox-srv001.database.windows.net',
  port '1433',
  user 'gizmoboxadmin',
  password 'admin@123'
)
```

### 2. Create foreign catalog - azure_sql_gizmobox_db_catalog

```sql
create foreign catalog if not exists azure_sql_gizmobox_db_catalog using connection azure_sql_gizmobox_db_conn
options(
  database 'gizmobox-db'
)
```

```sql
show catalogs;
use catalog azure_sql_gizmobox_db_catalog;
show schemas; 
use schema dbo;
show tables;
select * from refunds;
```

### We can use this table to join with other tables under unity catalog/Databricks Lakehouse

```sql
select * from gizmobox.bronze.payments p
join azure_sql_gizmobox_db_catalog.dbo.refunds r
on p.payment_id = r.payment_id;
```
