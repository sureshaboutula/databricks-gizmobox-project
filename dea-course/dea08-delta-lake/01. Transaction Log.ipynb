# Understanding Delta Lake Transaction Log
Understand the cloud storage directory structure behind Delta Lake Tables

### 0. Create a new schema under the demo catalog for this section of the course (delta lake)

```python
%sql
create external location if not exists dea_course_ext_dl_demo
  url 'abfss://demo@deacourseextdldev.dfs.core.windows.net/'
  with (storage credential dea_course_ext_sc)
```

```python
%fs ls 'abfss://demo@deacourseextdldev.dfs.core.windows.net/'
```

```python
%sql
create catalog if not exists demo
managed location 'abfss://demo@deacourseextdldev.dfs.core.windows.net/'
```

```python
%sql
create schema if not exists demo.delta_lake
  managed location 'abfss://demo@deacourseextdldev.dfs.core.windows.net/delta_lake'
```

### 1. Create a Delta Lake table

```python
%sql
create table if not exists demo.delta_lake.companies
  (company_name string,
  founded_date date,
  country string)
```

```python
%sql
describe extended demo.delta_lake.companies;
```

### 2. Insert Some Data

```python
%sql
insert into demo.delta_lake.companies
values ("Apple", "1976-04-01", "USA");
```

```python
%sql
select * from demo.delta_lake.companies;
```
