# 15. Query Orders Data as JSON Strings
1. Extract Top Level Column Values
2. Extract Array Elements
3. Extract Nested Column Values
4. CAST Column values to a Specific Data Type

```sql
select *
from gizmobox.bronze.v_orders;
```

### 1. Extract Top Level Column Values
Top Level Object : column_name:extraction_path

```sql
select value:order_id as order_id, value
from gizmobox.bronze.v_orders;
```

### 2. Extract Array Elements
To extract array elements - use the indices in brackets column_name:extraction_path[index]

```sql
select value:order_id as order_id, value:items as items, value
from gizmobox.bronze.v_orders;
```

```sql
select value:order_id as order_id, 
value:items[0] as item_1, 
value:items[1] as item_2, 
value
from gizmobox.bronze.v_orders;
```

### 3. Extract Nested Column Values
To Extract nested fields, use dot notation column_name:extraction_path[index].leaf_node

```sql
select value:order_id as order_id, 
value:items[0].item_id as item_1_item_id,
value:items[0] as item_1, 
value:items[1] as item_2, 
value
from gizmobox.bronze.v_orders;
```

### 4. CAST Column values to a Specific Data Type

```sql
select value:order_id :: integer as order_id, 
value:items[0].item_id :: Integer as item_1_item_id,
value:items[0] as item_1, 
value:items[1] as item_2, 
value
from gizmobox.bronze.v_orders;
```
