# Extract Data From the Payments Files - CSV
1. Read Payments File
2. Create Payments Table in Bronze Schema

### List the files from the Payments Folder

```python
%fs ls 'abfss://gizmobox@deacourseextdldev.dfs.core.windows.net/landing/external_data/payments'
```

### 1. Read Payments file

```python
payments_schema = 'payment_id integer, order_id integer, payment_timestamp timestamp, payment_status integer, payment_method string'
```

```python
payments_df = (
    spark.read.format("csv")
        .option("delimiter",",")
        .schema(payments_schema)
        .load('abfss://gizmobox@deacourseextdldev.dfs.core.windows.net/landing/external_data/payments')
)
display(payments_df)
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType

py_payments_schema = StructType([
    StructField('payment_id', IntegerType()),
    StructField('order_id', IntegerType()),
    StructField('payment_timestamp', TimestampType()),
    StructField('payment_status', IntegerType()),
    StructField('payment_method', StringType())
])
```

```python
payments_df = (
    spark.read.format("csv")
        .option("delimiter",",")
        .schema(py_payments_schema)
        .load('abfss://gizmobox@deacourseextdldev.dfs.core.windows.net/landing/external_data/payments')
)
display(payments_df)
```

### 2. Create Payments Table in Bronze Schema

```python
payments_df.writeTo("gizmobox.bronze.py_payments").createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_payments;
```
