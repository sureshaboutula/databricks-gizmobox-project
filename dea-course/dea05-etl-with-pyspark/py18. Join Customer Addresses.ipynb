# Join Customer and Address
Join Customers Data with Address Data to create customer_address table which contains the address of each customer on the same record

```python
customers_df = spark.table('gizmobox.silver.py_customers')
display(customers_df)
```

```python
addresses_df = spark.table('gizmobox.silver.py_addresses')
display(addresses_df)
```

```python
customer_addresses_df = customers_df.join(addresses_df, "customer_id", 'inner') \
    .select('customer_id', 
            'customer_name', 
            'email', 
            'date_of_birth', 
            'member_since', 
            'telephone', 
            'shipping_address_line_1', 'shipping_city', 'shipping_state', 'shipping_postcode', 
            'billing_address_line_1', 'billing_city', 'billing_state', 'billing_postcode')
display(customer_addresses_df)
```

```python
customer_addresses_df.writeTo('gizmobox.gold.py_customer_address').createOrReplace()
```
