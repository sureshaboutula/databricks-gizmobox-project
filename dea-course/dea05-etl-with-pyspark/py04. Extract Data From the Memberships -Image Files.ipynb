# Extract Data From the Memberships - Image Files
1. Query Memberships File using binaryFile format
2. Create Memberships Table in Bronze schema

### 1. Query Memberships File using binaryFile format

```python
%fs ls '/Volumes/gizmobox/landing/operational_data/memberships'
```

```python
memberships_df = spark.read.format("binaryFile").load("/Volumes/gizmobox/landing/operational_data/memberships/*")
display(memberships_df)
```

### 2. Create Memberships Table in Bronze Schema

```python
memberships_df.writeTo("gizmobox.bronze.py_memberships").createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_memberships;
```
