# Transform Memberships Data
1. Extract customer_id from the file_path
2. Write transformed Data to the Silver Schema

```python
memberships_df = spark.table('gizmobox.bronze.py_memberships')
display(memberships_df)
```

### 1. Extract customer_id from the file_path

```python
from pyspark.sql.functions import *
extracted_memberships_df = memberships_df.select(
    regexp_extract('path', r'.*/([0-9]+)\.png$', 1).alias('customer_id'),
    col('content').alias('membership_card')
    )
display(extracted_memberships_df)
```

### 2. Write transformed Data to the Silver Schema

```python
extracted_memberships_df.writeTo('gizmobox.silver.py_memberships').createOrReplace()
```
