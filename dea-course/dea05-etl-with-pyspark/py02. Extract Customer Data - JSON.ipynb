## Extract Data From the Customers JSON file
1. Query single file
2. Query list of files using wildcard charcters
3. Query all the files in the folder
4. Select File Metadata
5. Cretae Table in Bronze Schema

```python
%fs ls /Volumes/gizmobox/landing/operational_data/customers/
```

### 1. Query single JSON File

```python
customer_df = spark.read.format("json").load("/Volumes/gizmobox/landing/operational_data/customers/customers_2024_10.json")
display(customer_df)
```

### 2. Query multiple JSON files

```python
customer_df = spark.read.format("json").load("/Volumes/gizmobox/landing/operational_data/customers/customers_2024_*.json")
display(customer_df)
```

### 3. Query all the JSON files in the Folder

```python
customer_df = spark.read.format("json").load("/Volumes/gizmobox/landing/operational_data/customers/")
display(customer_df)
```

```python
df = spark.read.json("/Volumes/gizmobox/landing/operational_data/customers/")
display(df)
```

### 4. Select File MetaData

```python
df.select("_metadata.file_path", "*")
```

```python
df_with_metadata = df.select("_metadata.file_path", "*")
display(df_with_metadata)
```

### 5. Create Table in Bronze Schema

```python
df_with_metadata.write.format("delta").mode("overwrite").saveAsTable("gizmobox.bronze.py_customers")
```

```python
%sql
select * from gizmobox.bronze.py_customers;
```

```python
df_with_metadata.writeTo("gizmobox.bronze.py_customers").createOrReplace()
```
