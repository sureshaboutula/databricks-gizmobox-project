## Extract Data From the Orders JSON File
1. Query Orders file using JSON Format
2. Query Orders file using Text Format
3. Create Orders Table in Bronze Schema 

### 1. Query Orders file using JSON Format

```python
orders_df = spark.read.json("/Volumes/gizmobox/landing/operational_data/orders")
display(orders_df)
```

### 2. Query Orders file using TEXT Format

```python
orders_df = spark.read.text("/Volumes/gizmobox/landing/operational_data/orders")
display(orders_df)
```

### 3. Create Orders Table in Bronze layer

```python
orders_df.writeTo("gizmobox.bronze.py_orders").createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_orders;
```
