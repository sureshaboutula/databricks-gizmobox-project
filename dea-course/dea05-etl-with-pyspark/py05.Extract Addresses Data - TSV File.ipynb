# Extract Data From the Address Files
1. Read all the Address Files
2. Create Addresses Table in the Bronze schema

### 1. Read all the Address Files

```python
address_df = (spark.read.format('csv')
.option('header', 'true')
.option('delimiter', '\t')
.load('/Volumes/gizmobox/landing/operational_data/addresses'))
display(address_df)
```

### 2. Create Addresses Table in the Bronze schema

```python
address_df.writeTo("gizmobox.bronze.py_addresses").createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_addresses;
```
