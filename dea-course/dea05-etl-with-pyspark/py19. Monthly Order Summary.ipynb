# Monthly Order Summary
For each of the customer, produce the following summary per month
1. total orders
2. total items bought
3. total amount spent

```python
orders_df = spark.table('gizmobox.silver.py_orders')
display(orders_df)
```

```python
from pyspark.sql.functions import *
order_summary_df = orders_df.withColumn("order_month", date_format("transaction_timestamp", "yyyy-MM")) \
    .groupBy('order_month', 'customer_id') \
        .agg(
            countDistinct('order_id').alias('number_of_orders'),
            sum('quantity').alias('total_items_bought'),
            sum(col('price') * col('quantity')).alias('total_amount_spent')
            )
display(order_summary_df)
```

```python
display(order_summary_df.where(order_summary_df.customer_id == 9179))
```

```python
order_summary_df.writeTo('gizmobox.gold.py_order_summary_monthly').createOrReplace()
```
