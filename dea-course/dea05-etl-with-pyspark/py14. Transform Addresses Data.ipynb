# Transform Addresses Data
1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address
2. Write Transformed Data to Silver Schema

```python
addresses_df = spark.table('gizmobox.bronze.py_addresses')
display(addresses_df)
```

### 1. Create one record for each customer with 2 sets of address column, 1 for shipping and 1 for billing address

```python
from pyspark.sql.functions import *
pivoted_addresses_df = addresses_df.groupBy('customer_id').pivot('address_type', ['shipping', 'billing']) \
  .agg(
    max('address_line_1').alias('address_line_1'), 
    max('city').alias('city'), 
    max('state').alias('state'), 
    max('postcode').alias('postcode')
    )
display(pivoted_addresses_df)
```

### 2. Write Transformed Data to Silver Schema

```python
pivoted_addresses_df.writeTo('gizmobox.silver.py_addresses').createOrReplace()
```
