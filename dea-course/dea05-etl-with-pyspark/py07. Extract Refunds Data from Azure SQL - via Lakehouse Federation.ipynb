# Extract Data from the Refunds SQL Table
1. Read Refunds Data via JDBC
2. Create Refunds Table in Bronze Schema 

### 1. Read Refunds Data via JDBC

```python
refunds_df = (
    spark.read.format('jdbc')
        .option('url', 'jdbc:sqlserver://gizmobox-srv001.database.windows.net:1433;database=gizmobox-db')
        .option('dbtable', 'refunds')
        .option('user', 'gizmoboxadmin')
        .option('password', 'admin@123')
        .load()
)
display(refunds_df)
```

### 2. Create Refunds Table in Bronze Schema

```python
refunds_df.writeTo("gizmobox.bronze.py_refunds").createOrReplace()
```

```python
%sql
select * from gizmobox.bronze.py_refunds
```
